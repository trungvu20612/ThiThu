```python
import cv2
import numpy as np
import random

# Đọc ảnh gốc
image = cv2.imread('a.jpg')

# 1. Áp dụng mean filter
mean_filtered = cv2.blur(image, (5, 5))
cv2.imwrite('a_mean.jpg', mean_filtered)

# 2. Áp dụng filter xác định biên
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges = cv2.Laplacian(gray_image, cv2.CV_64F)
edges = cv2.convertScaleAbs(edges)
cv2.imwrite('a_edge.jpg', edges)

# 3. Đổi màu ngẫu nhiên
random_indices = list(np.random.permutation(3))
random_colored = image[:, :, random_indices]
cv2.imwrite('a_random_color.jpg', random_colored)

# 4. Chuyển sang HSV và tách kênh
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_image)
cv2.imwrite('a_hue.jpg', h)
cv2.imwrite('a_saturation.jpg', s)
cv2.imwrite('a_value.jpg', v)

print("Xử lý ảnh hoàn tất.")
```

```python
import cv2
import numpy as np
import random
import os

# Danh sách ảnh
image_files = ['image1.jpg', 'image2.jpg', 'image3.jpg']

# Hàm xử lý cho từng phương pháp
def inverse_transform(image):
    return 255 - image

def gamma_correction(image):
    gamma = random.uniform(0.5, 2.0)
    invGamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** invGamma * 255 for i in range(256)]).astype("uint8")
    return cv2.LUT(image, table), gamma

def log_transform(image):
    c = random.uniform(1.0, 5.0)
    image_float = image.astype(float)
    image_log = c * np.log1p(image_float)
    image_log = np.clip(image_log * 255 / np.max(image_log), 0, 255).astype(np.uint8)
    return image_log, c

def histogram_equalization(image):
    if len(image.shape) == 3:
        image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
        image_yuv[:, :, 0] = cv2.equalizeHist(image_yuv[:, :, 0])
        return cv2.cvtColor(image_yuv, cv2.COLOR_YUV2BGR)
    else:
        return cv2.equalizeHist(image)

def contrast_stretching(image):
    min_val = random.randint(0, 100)
    max_val = random.randint(150, 255)
    image_stretched = np.clip((image - min_val) * (255 / (max_val - min_val)), 0, 255).astype(np.uint8)
    return image_stretched, min_val, max_val

def adaptive_histogram_equalization(image):
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    if len(image.shape) == 3:
        lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
        lab[:, :, 0] = clahe.apply(lab[:, :, 0])
        return cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)
    else:
        return clahe.apply(image)

# Hàm xử lý menu
def process_images(method):
    for idx, file in enumerate(image_files, 1):
        image = cv2.imread(file)
        if image is None:
            print(f"Không tìm thấy ảnh: {file}")
            continue

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        if method == 'I':
            result = inverse_transform(image)
            cv2.imwrite(f'output_inverse_{idx}.jpg', result)

        elif method == 'G':
            result, gamma = gamma_correction(image)
            print(f"Gamma ảnh {idx}: {gamma:.2f}")
            cv2.imwrite(f'output_gamma_{idx}.jpg', result)

        elif method == 'L':
            result, c = log_transform(image)
            print(f"Hệ số log ảnh {idx}: {c:.2f}")
            cv2.imwrite(f'output_log_{idx}.jpg', result)

        elif method == 'H':
            result = histogram_equalization(image)
            cv2.imwrite(f'output_histogram_{idx}.jpg', result)

        elif method == 'C':
            result, min_v, max_v = contrast_stretching(gray)
            print(f"Contrast stretch ảnh {idx}: min={min_v}, max={max_v}")
            cv2.imwrite(f'output_contrast_{idx}.jpg', result)

        elif method == 'A':
            result = adaptive_histogram_equalization(image)
            cv2.imwrite(f'output_adaptive_{idx}.jpg', result)

# Menu chính
def show_menu():
    print("Chọn phương pháp biến đổi ảnh:")
    print("Phím I: Image inverse transformation")
    print("Phím G: Gamma-Correction")
    print("Phím L: Log Transformation")
    print("Phím H: Histogram Equalization")
    print("Phím C: Contrast Stretching")
    print("Phím A: Adaptive Histogram Equalization")

    method = input("Nhập phím: ").strip().upper()
    if method in ['I', 'G', 'L', 'H', 'C', 'A']:
        process_images(method)
        print("Đã xử lý xong và lưu ảnh.")
    else:
        print("Phím không hợp lệ.")

if __name__ == "__main__":
    show_menu()
```

```python
import cv2
import numpy as np
import random

# 1. Tăng kích thước ảnh colorful-ripe-tropical-fruits.jpg thêm 30px mỗi chiều
def resize_add_margin(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Lỗi đọc ảnh: {image_path}")
        return
    height, width = image.shape[:2]
    resized = cv2.resize(image, (width + 30, height + 30))
    cv2.imwrite("output_colorful_resized.jpg", resized)

# 2. Xoay ảnh quang-ninh.jpg 45 độ theo chiều kim đồng hồ và lật ngang
def rotate_and_flip(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Lỗi đọc ảnh: {image_path}")
        return
    height, width = image.shape[:2]
    center = (width // 2, height // 2)
    
    # Ma trận xoay
    matrix = cv2.getRotationMatrix2D(center, -45, 1.0)
    rotated = cv2.warpAffine(image, matrix, (width, height))
    
    # Lật ngang
    flipped = cv2.flip(rotated, 1)
    cv2.imwrite("output_quangninh_rotated_flipped.jpg", flipped)

# 3. Phóng to ảnh pagoda.jpg lên 5 lần + Gaussian blur kernel 7x7
def upscale_and_blur(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Lỗi đọc ảnh: {image_path}")
        return
    height, width = image.shape[:2]
    resized = cv2.resize(image, (width * 5, height * 5), interpolation=cv2.INTER_LINEAR)
    blurred = cv2.GaussianBlur(resized, (7, 7), 0)
    cv2.imwrite("output_pagoda_upscaled_blurred.jpg", blurred)

# 4. Áp dụng công thức tăng độ sáng & tương phản cho pagoda.jpg
def contrast_brightness_adjust(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Lỗi đọc ảnh: {image_path}")
        return

    alpha = random.uniform(0.5, 2.0)     # contrast factor
    beta = random.randint(-50, 50)       # brightness offset

    print(f"Áp dụng contrast α = {alpha:.2f}, brightness β = {beta}")

    # Áp dụng công thức: I_out = α * I_in + β (và clip giá trị)
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

    cv2.imwrite("output_pagoda_brightness_contrast.jpg", adjusted)

# -------------------------------
# Thực hiện toàn bộ yêu cầu đề bài
# -------------------------------
resize_add_margin("colorful-ripe-tropical-fruits.jpg")
rotate_and_flip("quang-ninh.jpg")
upscale_and_blur("pagoda.jpg")
contrast_brightness_adjust("pagoda.jpg")

print("Đã xử lý xong toàn bộ ảnh.")
```
